```python
image = cv2.imread('C:/Users/choye/Python_Application/image/dark.png')
```

```python
# 흑백 이미지
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
_, thresh = cv2.threshold(image_gray, 16, 255, cv2.THRESH_BINARY)

# 이미지 출력
plt.imshow(image_gray, 'gray')
plt.show()
```

# 히스토그램

```python
# 흑백
img_hist = cv2.calcHist(image_gray,[0],None,[256],[0,256])
plt.plot(img_hist)
plt.xlim([0,256])
# plt.ylim([0,150])
plt.show()
```

```python
# 컬러
image_bgr = image

color = ('b','g','r')
for i, col in enumerate(color):
    img_hist = cv2.calcHist(image_bgr,[i],None,[256],[0,256])
    plt.plot(img_hist, color=col)
    plt.xlim([0,256])
plt.show()
```

# Edge 추출

```python
# 흑백
image_gray = cv2.imread('C:/Users/choye/Python_Application/image/sample-color.png', cv2.IMREAD_GRAYSCALE)

# 캐니 경계선 감지기
# 중간 픽셀 강도에서 위아래 1 표준 편차 떨어진 값을 임계값으로 지정합니다.
image_canny = cv2.Canny(image_gray, 70, 130)

plt.imshow(image_canny, cmap='gray')
plt.show()
```

# image blending
cv2.addWeighted 함수의 인자가 (imgA, a, imgB, b, c)라고 한다면 산술식은 다음과 같습니다.

result = imgA + a + imgB + b + c

```python
img1 = cv2.imread('C:/Users/choye/Python_Application/image/img1.jpg')
img2 = cv2.imread('C:/Users/choye/Python_Application/image/img2.jpg')

# result = imgA + a + imgB + b + c
blending = cv2.addWeighted(img1, 0.7, img2, 0.3, 50)

# 이미지를 출력
plt.imshow(cv2.cvtColor(blending, cv2.COLOR_BGR2RGB))
plt.show()
```

# image insert

```python
# Load two image
img1 = cv2.imread('C:/Users/choye/Python_Application/image/img1.jpg')
img2 = cv2.imread('C:/Users/choye/Python_Application/image/logo.jpg')

# 삽입할 이미지의 rows, cols, channels 정보
rows1, cols1, channels1 = img1.shape
rows2, cols2, channels2 = img2.shape

# 합쳐진 이미지를 원본 이미지에 추가.
# img1[0:rows2, 0:cols2] = img2             # 좌측상단
img1[(rows1-rows2):rows1, 0:cols2] = img2   # 좌측하단

# 이미지를 출력
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.show()
```

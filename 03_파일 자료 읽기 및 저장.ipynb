## 공적마스크 판매량 및 재고
http://youngho.iwinv.net/data/link.txt

```python
url_json = 'https://8oi9s0nnth.apigw.ntruss.com/corona19-masks/v1/storesByAddr/json?address=%EB%8C%80%EC%A0%84%EA%B4%91%EC%97%AD%EC%8B%9C%20%EB%8C%80%EB%8D%95%EA%B5%AC'
data_mask = pd.read_json(url_json, orient='index')
data_mask
```

```python
json_mask= data_mask.loc['stores', 0]
#키를 기용할때는 인덱스 행이름, 열이름
type(json_mask)
```

```python
df_mask = pd.DataFrame(json_mask)
df_mask
#lat = 위도 경도 정보,stat = 얼마나 있냐
```

```python
df_mask.to_csv('mask.csv', encoding='ms949')
#csv로 변환 - 글자 다 깨짐 엉엉
#그래서 인코딩 해줌
```

```python
url = 'https://8oi9s0nnth.apigw.ntruss.com/corona19-masks/v1/storesByAddr/json?address='
addr = '대전광역시 동구'

## url encoding 하는 라이브러리 불러오기
import urllib.parse as up

url_json = url + up.quote(addr)#인코딩해주는 함수
data_mask = pd.read_json(url_json, orient='index')
data_mask
```

```python
json_mask= data_mask.loc['stores', 0]
#키를 기용할때는 인덱스 행이름, 열이름
type(json_mask)
```

```python
df_mask = pd.DataFrame(json_mask)
df_mask
#lat = 위도 경도 정보,stat = 얼마나 있냐
```

```python
df_mask.to_csv('mask.csv', encoding='ms949')
```

## SQLite - iris.sqlite
* http://youngho.iwinv.net/data/iris.sqlite

```python
from tkinter.filedialog import askopenfilename
```

```python
database = askopenfilename()
database
```

```python
import sqlite3
```

```python
# DB connect
conn = sqlite3.connect(database)
```

```python
## Table list
query = '''
select *
from sqlite_master
where type='table'
'''
pd.read_sql(query, conn)
```

```python
### Exqute query
query = '''
select *
from Iris
'''
pd.read_sql(query, conn)
```

```python
### Disconnect
conn.close()
```

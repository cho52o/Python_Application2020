```python
import pandas as pd
```

## Read data - titanic.csv
* http://youngho.iwinv.net/data/titanic.csv

## 열 추가

```python
data_file = 'http://youngho.iwinv.net/data/titanic.csv'
data_raw = pd.read_csv(data_file)
data_raw
```

```python
data_raw['fare_round'] = 0
data_raw
```

```python
data_raw['fare_round'] = round(data_raw.fare, 0)
data_raw
```

## 열 제거

```python
### data_raw
data_raw.columns
```

```python
### 메모리에 독립적으로 저장
data_use = data_raw.copy()

### Drop column
col_names = ['Unnamed: 0', 'fare_round']
data_use = data_use.drop(col_names, axis=1)
data_use
```

## 조건 필터

```python
data = data_use[(data_use.survived == 1)]
data
```

```python
data = data_use[(data_use.survived == 1) & (data_use.sex == 'female')]
data
```

## 열 선택

```python
col_names = ['age', 'fare', 'cabin']
data_sub = data[col_names].copy()
data_sub
```

## 결측값 처리 - NaN(NA), Null
 - 무시
 - 제거
 - 대체

```python
## 제거
data_sub.dropna()
```

```python
## 대체
data_sub.fillna('NA')
```

```python
## 평균 대체
data_sub.iloc[0:4, 0:2] = None
data_sub
```

```python
data_sub.fillna(data_sub.mean())
```

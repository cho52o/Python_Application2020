# 다중 클래스 분류기
* 3개 이상의 클래스(범주)가 있는 경우

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
```

```python
# data loed
iris = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/iris.csv')
iris
```

```python
### 설명변수(x)와 반응변수(y)
x = iris[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']]
y = iris['Species']
```

```python
### Train & Test data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.4)
```

```python
### logistic regression
# multi_class={'auto', 'ovr', 'multinomial'}, default='auto'
logistic = LogisticRegression(multi_class='ovr')
model = logistic.fit(x_train, y_train)
```

```python
### 예측
y_pred = model.predict(x_test)
y_pred
```

```python
### 교차표
pd.crosstab(y_test, y_pred)
```

```python
### accuracy
model.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
# data loed
house = pd.read_csv('http://youngho.iwinv.net/data/house_price_prediction.csv')
house
```

```python
house.describe()
```

```python
### 설명변수(x)와 반응변수(y)
x = house[['bedrooms','sqft_living','view','condition','yr_built']]
y = house['price_cat']
```

```python
### Train & Test data
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
### logistic regression
# multi_class={'auto', 'ovr', 'multinomial'}, default='auto'
logistic = LogisticRegression(multi_class='ovr')
model = logistic.fit(x_train, y_train)
```

```python
### 예측
y_pred = model.predict(x_test)
y_pred
```

```python
### 교차표
pd.crosstab(y_test, y_pred)
```

```python
### accuracy
model.score(x_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```
